# Parsing the preprocessing data
## Two variants

```python
ad2 =autodock2.filter(['Label','Molecule'], axis=1)
ad2=ad2.rename(columns={'Label':'Label_ad','Molecule':'Molecule_ad'})

vin2=vina2.filter(['Label','Molecule'], axis=1)
vin2=vin2.rename(columns={'Label':'Label_vin','Molecule':'Molecule_vin'})

smin2=smina2.filter(['Label','Molecule'], axis=1)
smin2=smin2.rename(columns={'Label':'Label_smin','Molecule':'Molecule_smin'})

plan2=plants2.filter(['Label','Molecule'], axis=1)
plan2=plan2.rename(columns={'Label':'Label_plan','Molecule':'Molecule_plan'})

rdoc2=rdock2.filter(['Label','Molecule'], axis=1)
rdoc2=rdoc2.rename(columns={'Label':'Label_rdoc','Molecule':'Molecule_rdoc'})
```

```python
total_2=pd.concat([ad2.reset_index(drop=True), vin2.reset_index(drop=True), smin2.reset_index(drop=True), plan2.reset_index(drop=True), rdoc2.reset_index(drop=True)],axis=1)
```

```python
total_2.head(10)
```

```python
total_2.head(n=20).to_csv('total_2.csv')
```

### ECR function two variants

```python
numeros2 = list(index_2.loc[:,"Molecule"].values)
```

```python
cons2_ad = list(total_2.loc[:, "Molecule_ad"].values)
ind2_ad = [cons2_ad.index(num)+1 for num in numeros2 if num in cons2_ad]

cons2_vin = list(total_2.loc[:, "Molecule_vin"].values)
ind2_vin = [cons2_vin.index(num)+1 for num in numeros2 if num in cons2_vin]

cons2_smin = list(total_2.loc[:, "Molecule_smin"].values)
ind2_smin = [cons2_smin.index(num)+1 for num in numeros2 if num in cons2_smin]

cons2_plan = list(total_2.loc[:, "Molecule_plan"].values)
ind2_plan = [cons2_plan.index(num)+1 for num in numeros2 if num in cons2_plan]

cons2_rdoc = list(total_2.loc[:, "Molecule_rdoc"].values)
ind2_rdoc = [cons2_rdoc.index(num)+1 for num in numeros2 if num in cons2_rdoc]
```

```python
ecr2 = pd.DataFrame(
    {'ad2': ind2_ad,
     'vin2': ind2_vin,
     'smin2': ind2_smin,
     'plan2': ind2_plan,
     'rdoc2': ind2_rdoc
    })
```

```python
ecr2
```

```python
ecr2=ecr2.apply(lambda x: np.exp(-x/20)/20, axis=1)
ecr2["ECR"] = ecr2.sum(axis=1)
ecr2["Molecule"]=list(range(1, 156))
```

```python
ecr2
```

```python
labelecr_2=index_2.drop_duplicates(subset=['Label']).drop(columns=['Molecule'])
labelecr_2["Molecule"]=list(range(1, 156))
```

```python
ecr2=ecr2[['ECR','Molecule']]
ecr2=labelecr_2.merge(ecr2, on='Molecule', how='left')
```

```python
ecr2=ecr2.sort_values(by=['ECR'],ascending=False)
ecr2['Molecule_par']=ecr2['Molecule'].apply(lambda x: x*2)
ecr2['Molecule_impar']=ecr2['Molecule'].apply(lambda x: x*2-1)
```

```python
ecr2.head(n=20)
```

```python
ecr2.to_csv('ecr2N.csv')
```
