## Parsing Output CSV Autodock-GPU

```python
##Importing packages

import pandas as pd
import numpy as np
```

```python
##Reading file

df = pd.read_csv('output.csv', delimiter="@",names=['Data1', 'Data2'])
df.head(n=20)
```

```python
df=df.ffill(axis = 0) 
df.head(n=5)
```

```python
df[['Output','Value']] = df.Data2.str.split("=",expand=True,)
df.drop(['Data2'], axis=1)
```

```python
df_pivot=pd.pivot_table(df, index = ["Data1"],values = ["Value"], columns = "Output", aggfunc='first').reset_index()
```
