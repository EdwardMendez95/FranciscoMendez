# Parsing pharmacophore output (Pharmagist)
### Three ligands N-termini

## Two variants

```python
##Importing packages

import pandas as pd
import numpy as np
import math
```

```python
##Importing dataset
VS_pharma_2=pd.read_csv('scores_1.pha_db_2_var.mol2.csv',header=None,names=["Index", "Score","Aromatic","Hydrophobic","Donors","Aceptors", "Negatives","Molecule"]) 
index_2=pd.read_csv('indice_2.csv')  
index_2['Molecule'] = list(range(1, 311))
index_2.Molecule=index_2.Molecule.astype('int64')
index_2
```

```python
VS_pharma_2['Molecule']=VS_pharma_2['Molecule'].str.replace("2_var_", "")
VS_pharma_2['Molecule']=VS_pharma_2['Molecule'].str.replace(".mol2", "")
```

```python
VS_pharma_2.Molecule=VS_pharma_2.Molecule.astype('int64')
```

```python
VS_pharma_2
```

```python
VS_pharma_2=index_2.merge(VS_pharma_2, on='Molecule', how='left').sort_values(by="Index")
```

```python
VS_pharma_2=VS_pharma_2.drop_duplicates(subset=['Label'])
VS_pharma_2.Molecule=VS_pharma_2.Molecule.div(2).apply(np.ceil).astype('int64')
```

```python
VS_pharma_2.head(20)
```

```python
VS_pharma_2.to_csv('VS_2N.csv')
```
