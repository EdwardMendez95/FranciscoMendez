# Final results
# C_terminal

```python
##Importing packages

import pandas as pd
import numpy as np
import math
```

## Ligand index

```python
indice=pd.read_csv('indice_2.csv')  
indice=indice.drop_duplicates(subset=['Label'])
indice['Molecule'] = list(range(1, 156))
```

## Importing data (Two variants)

```python
farma_2c=pd.read_csv('farma_2C.csv',sep=',')  
docking_2c=pd.read_csv('dockingecr2C.csv',sep=',')  
qsar2=pd.read_csv('Output_ECR_2_final.csv',sep=',') 
```

```python
farma_2c=farma_2c.drop(columns=["Number", "Index", "Score", "Aromatic", "Hydrophobic", "Donors", "Aceptors", "Negatives"])
farma_2c=farma_2c.rename(columns={'Label': 'Label_farma2c','Molecule':'Molecule_farma2c'})
farma_2c
```

```python
docking_2c=docking_2c.drop(columns=["Unnamed: 0", "ECR", "Molecule_par", "Molecule_impar"])
docking_2c=docking_2c.rename(columns={'Label': 'Label_docking2c','Molecule':'Molecule_docking2c'})
docking_2c
```

```python
qsar2=qsar2.drop(columns=["ECR"])
qsar2=qsar2.rename(columns={'Index': 'Molecule_qsar2','Molecule':'Label_qsar2'})
qsar2
```

```python
C2_lista = [farma_2c, docking_2c, qsar2]
```

```python
C2 = pd.concat(C2_lista, axis=1)
C2
```

### ECR 

```python
numeros = list(indice.loc[:,"Molecule"].values)
```

```python
q_2c = list(C2.loc[:, "Molecule_qsar2"].values)
q_2c_ind = [q_2c.index(num)+1 for num in numeros if num in q_2c]

d_2c = list(C2.loc[:, "Molecule_docking2c"].values)
d_2c_ind = [d_2c.index(num)+1 for num in numeros if num in d_2c]

f_2c = list(C2.loc[:, "Molecule_farma2c"].values)
f_2c_ind = [f_2c.index(num)+1 for num in numeros if num in f_2c]
```

```python
ECR_2C = pd.DataFrame(
    {'q_2c': q_2c_ind,
     'd_2c': d_2c_ind,
     'f_2c': f_2c_ind
    })
```

```python
ECR_2C
```

```python
ECR_2C=ECR_2C.apply(lambda x: np.exp(-x/20)/20, axis=1)
ECR_2C["ECR"] = ECR_2C.sum(axis=1)
ECR_2C["Molecule"]=list(range(1, 156))
ECR_2C=ECR_2C.sort_values(by=['ECR'],ascending=False)
```

```python
ECR_2C
```

```python
ECR_2C.to_csv('ECR_2C.csv')
```
